## Data Overview:

### Dataset Description:
The dataset used for this analysis comprises information about employee performance and various attributes related to their work environment, job roles, and personal demographics. The dataset includes **1,200 records** and **28 features**. The features are categorized as follows:

- **Categorical Features**:
  - `EmpNumber`: Unique identifier for each employee (object)
  - `Gender`: Gender of the employee (object)
  - `EducationBackground`: Educational background of the employee (object)
  - `MaritalStatus`: Marital status of the employee (object)
  - `EmpDepartment`: Department in which the employee works (object)
  - `EmpJobRole`: Job role of the employee (object)
  - `BusinessTravelFrequency`: Frequency of business travel (object)
  - `OverTime`: Whether the employee works overtime (object)
  - `Attrition`: Whether the employee has left the company (object)

- **Numerical Features**:
  - `Age`: Age of the employee (int64)
  - `DistanceFromHome`: Distance between employee's home and workplace (int64)
  - `EmpEducationLevel`: Education level of the employee (int64)
  - `EmpEnvironmentSatisfaction`: Satisfaction with the work environment (int64)
  - `EmpHourlyRate`: Hourly rate of pay (int64)
  - `EmpJobInvolvement`: Involvement level in job (int64)
  - `EmpJobLevel`: Job level of the employee (int64)
  - `EmpJobSatisfaction`: Satisfaction with the job role (int64)
  - `NumCompaniesWorked`: Number of companies the employee has worked for (int64)
  - `EmpLastSalaryHikePercent`: Percentage of last salary hike (int64)
  - `EmpRelationshipSatisfaction`: Satisfaction with work relationships (int64)
  - `TotalWorkExperienceInYears`: Total work experience in years (int64)
  - `TrainingTimesLastYear`: Number of training times last year (int64)
  - `EmpWorkLifeBalance`: Work-life balance rating (int64)
  - `ExperienceYearsAtThisCompany`: Years of experience at the current company (int64)
  - `ExperienceYearsInCurrentRole`: Years of experience in the current role (int64)
  - `YearsSinceLastPromotion`: Years since the last promotion (int64)
  - `YearsWithCurrManager`: Years with the current manager (int64)
  - `PerformanceRating`: Performance rating of the employee (int64)

### Data Source:

##### 3. Exploratory Data Analysis (EDA):
- Department-wise Performance Analysis.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Countplot of performance ratings per department.
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='EmpDepartment', hue='PerformanceRating', palette='Set2')
plt.title('Department-wise Performance Rating Distribution')
plt.xlabel('Department')
plt.ylabel('Count of Employees')
plt.xticks(rotation=45)
plt.show()
```

- ***What it does:*** This plot displays how employee performance ratings (ranging from 1 to 4) are spread across different departments in the company.

- ***How to read it:*** Each department is represented by several bars, one for each performance rating. The height of the bars shows how many employees in that department received a particular rating. Higher bars indicate more employees with that rating.

- ***Insights:***

- You can quickly see which departments have a larger number of high-performing employees (ratings 3 or 4) versus those with more low-performing employees (ratings 1 or 2).
- If a department has a significant number of employees with low performance ratings, it may signal a need for management attention or further analysis to address potential issues affecting productivity.

```python
# Calculate the average performance rating per department.
avg_performance = df.groupby('EmpDepartment')['PerformanceRating'].mean().sort_values(ascending=False)

# Plot average performance rating.
plt.figure(figsize=(10,6))
avg_performance.plot(kind='bar', color='skyblue')
plt.title('Average Performance Rating by Department')
plt.xlabel('Department')
plt.ylabel('Average Performance Rating')
plt.xticks(rotation=90)
plt.show()
```

- ***What it does:*** This plot highlights the average performance rating for employees within each department.

- ***How to read it:*** Each bar corresponds to a specific department, and the height of the bar reflects the average performance rating (on a scale from 1 to 4). Taller bars indicate higher average ratings, meaning better overall performance in that department.

- ***Insights:***

- Departments with higher average performance ratings (closer to 4) are performing well and likely have motivated and productive teams.
- Lower average ratings may signal departments that are facing challenges, such as low morale, lack of resources, or training gaps, and may require management intervention to improve performance.

```python
# Barplot of performance ratings per department.
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='EmpDepartment', y='PerformanceRating', palette='Set3', ci=None)
plt.title('Performance Rating Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Performance Rating')
plt.xticks(rotation=90)
plt.show()
```

- ***What it does:*** This bar plot shows how employee performance ratings are distributed within each department.

- ***How to read it:*** The bar itself represents the middle 50% of performance ratings, with the line inside showing the median rating (the value right in the middle). The "whiskers" extend to show the overall spread of ratings, while any dots outside the whiskers indicate outliersâ€”employees whose ratings are significantly different from the rest.

- ***Insights:***
- Departments with shorter boxes have less variation in employee performance, meaning their employees tend to perform consistently.
- Longer boxes indicate more variation in performance, which could point to inconsistencies in employee productivity or challenges that affect performance unevenly across the department.

```python
# Attrition rate by department.
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='EmpDepartment', hue='Attrition', palette='Set2')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Count of Employees')
plt.xticks(rotation=45)
plt.show()
```

- ***What it does:*** This plot shows how many employees have left each department (Attrition = "Yes") compared to those who have stayed (Attrition = "No").

- ***How to read it:*** Each department has two barsâ€”one for employees who have stayed and one for those who have left. A higher "Yes" bar means more people are leaving that department.

- ***Insights:***

- Departments with a higher number of employees leaving could be facing issues like low job satisfaction, poor management, or lack of growth opportunities.
- Pinpointing these departments helps focus on improving employee retention, potentially through better support, resources, or engagement strategies.

```python
# Performance rating by overtime status.
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='OverTime', y='PerformanceRating', palette='Set2', ci=None)
plt.title('Performance Rating vs OverTime')
plt.xlabel('OverTime')
plt.ylabel('Average Performance Rating')
plt.show()
```

- ***What it does:*** This plot compares the average performance rating of employees who work overtime (OverTime = "Yes") with those who donâ€™t (OverTime = "No").

- ***How to read it:*** Each bar shows the average performance score for both groups. A higher bar for "Yes" means employees who work overtime have better performance on average, while a lower bar indicates the opposite.

- ***Insights:***

- On the other hand, if their performance is higher, it could indicate that overtime is helping employees stay on top of their tasks or that more productive employees tend to work longer hours.

The line inside the box indicates the median rating, while the "whiskers" show the overall spread of ratings. Outliers (dots outside the whiskers) indicate employees whose performance significantly deviates from the norm.

```python
# Scatter plot of total work experience vs performance rating.
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='TotalWorkExperienceInYears', y='PerformanceRating', hue='EmpDepartment', palette='Set1')
plt.title('Total Work Experience vs Performance Rating')
plt.xlabel('Total Work Experience (Years)')
plt.ylabel('Performance Rating')
plt.show()
```

##### 4. Feature Importance Analysis:

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
# Convert categorical variables to numerical values using one-hot encoding.
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
# Define features and target.
X = df_encoded.drop('PerformanceRating', axis=1)
y = df_encoded['PerformanceRating']
```

```python
# Split the data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Train a Random Forest model to get feature importance.
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X, y)
```

```python
# Get feature importance.
importance = model.feature_importances_

# Create a DataFrame for features and their importance.
features = X.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': importance})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
```

```python
# Display the top 3 features.
print("Top 3 Features Affecting Performance:")
print(importance_df.head(3))

# Plot the feature importances.
plt.figure(figsize=(12, 8))
plt.barh(importance_df['Feature'][:3], importance_df['Importance'][:3], color='skyblue')
plt.xlabel('Importance')
plt.title('Top 3 Factors Affecting Employee Performance')
plt.gca().invert_yaxis()
plt.show()
```

##### 5. Predictive Modeling:

```python
# Initialize models.
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=0),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=0),
    'Support Vector Regressor': SVR(kernel='rbf')
}
```

```python
# Train models and evaluate.
results = {}
for name, model in models.items():
    # Train the model.
    model.fit(X_train, y_train)
    
    # Make predictions.
    y_pred = model.predict(X_test)
    
    # Evaluate the model.
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results[name] = {'MAE': mae, 'R2': r2}
```

```python
# Display the results.
results_df = pd.DataFrame(results).T
print("Model Performance:")
print(results_df)

# Plot model performance.
results_df[['MAE', 'R2']].plot(kind='bar', figsize=(14, 7))
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
# Encode categorical variables.
df_encoded = pd.get_dummies(df, drop_first=True)

# Convert the target variable to a classification problem.
# Assuming PerformanceRating > 3 is considered high, otherwise low.
df_encoded['PerformanceRating'] = df_encoded['PerformanceRating'].apply(lambda x: 1 if x > 3 else 0)
```

```python
# Define features and target.
X = df_encoded.drop('PerformanceRating', axis=1)
y = df_encoded['PerformanceRating']

# Split the data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initialize models.
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=0),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=0),
    'Support Vector Classifier': SVC()
}
```

```python
# Train models and evaluate.
results = {}
for name, model in models.items():
    # Train the model.
    model.fit(X_train, y_train)
    
    # Make predictions.
    y_pred = model.predict(X_test)
    
    # Evaluate the model.
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    results[name] = {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1}
```

```python
# Display the results.
results_df = pd.DataFrame(results).T
print("Model Performance:")
print(results_df)

# Plot model performance.
results_df[['Accuracy', 'Precision', 'Recall', 'F1 Score']].plot(kind='bar', figsize=(14, 7))
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```
